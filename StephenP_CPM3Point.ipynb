```python
import pandas as pd

simulationDF = tasksDF = pd.read_csv('/content/drive/MyDrive/IT262 - IT Project Management/StephenP_3PointCPM/simulationDataset.csv')

paths = ['ADEJ', 'BCDEJ', 'BCFGJ', 'BCFHJ', 'BI']
pathsDF = pd.DataFrame(columns=paths)

for index, row in simulationDF.iterrows():
  pathLength = []
  for path in paths:
    length = 0
    for character in path:
      length = length + row[character]

    pathLength.append(length)

  pathsDF.loc[len(pathsDF.index)] = pathLength

pathsDF['critical'] = pathsDF.idxmax(axis=1)
print(pathsDF.critical.value_counts())
```
