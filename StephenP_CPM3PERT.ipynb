```python
import pandas as pd
import tensorflow_probability as tfp
tfd = tfp.distributions

tasksDF = pd.read_csv('/content/drive/MyDrive/IT262 - IT Project Management/PERT-CPM.csv')
taskNames = tasksDF['tasks']

#creating simulation DataFrame
simulationDF = pd.DataFrame(columns=taskNames)

print(simulationDF)
for index, task in tasksDF.iterrows():
  dist = tfd.PERT(low=task['best'], peak=task['mostL'], high=task['worst'], temperature=4)
  simulationDF[task['tasks']] = dist.sample(1000)

print(simulationDF.head(2))

simulationDF.to_csv('/content/drive/MyDrive/IT262 - IT Project Management/simulationDataset.csv', index=False)
```
